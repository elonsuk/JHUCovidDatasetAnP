```python
import numpy as np
import pandas as pd

CC_PATH = r"C:\Users\offco\Documents\Dev_Projects\JHUCovidDatasetAnP\modified\confirmed\confirmed_korea, south.csv"
D_PATH = r"C:\Users\offco\Documents\Dev_Projects\JHUCovidDatasetAnP\modified\deaths\deaths_korea, south.csv"
cc = pd.read_csv(CC_PATH)
d = pd.read_csv(D_PATH)
```

```python
since100 = cc[cc["Korea, South"] >= 100].head(1).index[0]

cc['day_since100'] = [0 for _ in range(0, len(cc))]
cc['day_since100'][:since100] = [i for i in range(-since100,0)]
cc['day_since100'][since100+1:] = [i for i in range(1, (len(cc)-since100))]

cc.head(-1)
```

```python
loop = len(cc)
data = pd.DataFrame({"country": ["Korea, South" for _ in range(loop)], 
        "province": [np.nan for _ in range(loop)], 
        "date": cc["Date"], 
        "day_since100": cc["day_since100"], 
        "case_cnt": cc["Korea, South"], 
        "death_cnt": d["Korea, South"], 
        "new_hospitalization": [np.nan for _ in range(loop)], 
        "total_hospitalization": [np.nan for _ in range(loop)], 
        "new_icu": [np.nan for _ in range(loop)], 
        "total_icu": [np.nan for _ in range(loop)], 
        "new_recovered": [np.nan for _ in range(loop)], 
        "total_recovered": [np.nan for _ in range(loop)]})

data.head()
```

```python
SAVE_PATH = r"C:\Users\offco\Documents\Dev_Projects\JHUCovidDatasetAnP\3. DELPHI Analysis & Prediction\Demo\danger_map\processed"
PROVINCE = "None" if pd.isna(data["province"][0]) else data["province"][0]
FILE_NAME = ("\Cases " + data["country"][0] + " " + PROVINCE + ".csv").replace(" ", "_")
data.to_csv(SAVE_PATH + FILE_NAME, index=False)
```
